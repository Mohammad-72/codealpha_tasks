# Step 3: Data Cleaning

Although the dataset does not contain missing values detected by `isnull()`, the `TotalCharges` column contains 11 blank string values.

These blank values are converted into numeric format and removed to ensure accurate analysis.

```python
# Convert TotalCharges to numeric

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
# Check missing values again

df.isnull().sum()
```

```python
# Remove missing values

df = df.dropna()
```

# Step 6: Contract Type Analysis

In this section, we analyze how different contract types are related to customer churn. This helps identify which customer groups are more likely to leave the service.

```python
# Contract Distribution

df["Contract"].value_counts()
```

```python
# Contract Percentage

round(df["Contract"].value_counts(normalize=True) * 100, 2)
```

```python
plt.figure(figsize=(7,5))

df["Contract"].value_counts().plot(kind="bar")

plt.title("Contract Type Distribution")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.show()
```

```python
pd.crosstab(df["Contract"], df["Churn"])
```

```python
pd.crosstab(
    df["Contract"],
    df["Churn"],
    normalize="index"
) * 100
```

### Observation

- Customers with **Month-to-Month contracts** have the highest churn rate (approximately **42.71%**).
- Customers with **One-Year contracts** have a much lower churn rate (approximately **11.28%**).
- Customers with **Two-Year contracts** have the lowest churn rate (approximately **2.85%**).
- This indicates that customers with long-term contracts are significantly more likely to stay with the company.
- Businesses can reduce churn by encouraging customers to switch from month-to-month plans to longer-term contracts through discounts or loyalty programs.

# Step 7: Monthly Charges Analysis

This section analyzes the distribution of monthly charges and examines whether higher monthly charges are associated with customer churn.

- Monthly charges range from **18.25** to **118.75**.

- This suggests that higher service costs may contribute to customer churn.

# Step 8: Tenure Analysis

This section analyzes the relationship between customer tenure and churn. Tenure represents the number of months a customer has stayed with the company.

Understanding this relationship helps identify whether new customers are more likely to churn than long-term customers.


```python
# Summary statistics for tenure

df["tenure"].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["tenure"], bins=20)

plt.title("Distribution of Customer Tenure")

plt.xlabel("Tenure (Months)")

plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Churn", y="tenure", data=df)

plt.title("Tenure vs Customer Churn")

plt.xlabel("Churn")

plt.ylabel("Tenure (Months)")

plt.show()
```

```python
df.groupby("Churn")["tenure"].mean()
```

### Observation

- Customers who stayed with the company generally have a higher average tenure.
- Customers who churned tend to have a much shorter tenure.
- This indicates that new customers are more likely to leave the company than long-term customers.
- Improving the onboarding experience and early customer engagement may help reduce churn.

## Step 9: Internet Service Analysis

This section analyzes customer distribution based on different internet service types and their relationship with customer churn.

The objective is to identify whether internet service type influences customer retention and churn behavior.

```python
# Count customers by internet service type

internet_service_count = df["InternetService"].value_counts()

internet_service_count
```

```python
plt.figure(figsize=(8,5))

internet_service_count.plot(kind="bar")

plt.title("Customer Distribution by Internet Service")

plt.xlabel("Internet Service Type")

plt.ylabel("Number of Customers")

plt.xticks(rotation=0)

plt.show()
```

```python
# Internet Service vs Churn

internet_churn = pd.crosstab(
    df["InternetService"],
    df["Churn"]
)

internet_churn
```

```python
plt.figure(figsize=(8,5))

internet_churn.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Churn Distribution by Internet Service")

plt.xlabel("Internet Service Type")

plt.ylabel("Number of Customers")

plt.xticks(rotation=0)

plt.show()
```

```python
# Churn Rate Analysis
internet_churn_rate = (
    df.groupby("InternetService")["Churn"]
    .value_counts(normalize=True)
    .unstack()
)

internet_churn_rate
```

```python
plt.figure(figsize=(8,5))

internet_churn_rate["Yes"].plot(kind="bar")

plt.title("Churn Rate by Internet Service")

plt.xlabel("Internet Service Type")

plt.ylabel("Churn Rate")

plt.xticks(rotation=0)

plt.show()
```

## Observation

- Internet service type has an impact on customer churn behavior.
- Customers using Fiber optic service should be analyzed further because they may show different churn patterns.
- Customers without internet service generally have lower engagement with online services.
- Internet service information can help identify high-risk customers for retention campaigns.
