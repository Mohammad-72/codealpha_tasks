## Project Objective

The objective of this project is to extract data from a public website using Python and BeautifulSoup. The scraped data includes quotes, authors, and tags, which are stored in a custom dataset for further analysis and visualization.

This project demonstrates:
- Web Scraping using Python
- HTML Parsing
- Multi-page Data Extraction
- Creating a Custom CSV Dataset

```python
# Import required libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Step 1: Scraping Data from Website

In this step, data is collected from all 10 pages of the website using Requests and BeautifulSoup.

```python
# Create empty lists to store scraped data

all_quotes = []
all_authors = []
all_tags = []

# Loop through all 10 pages

for page in range(1,11):

    url = f"https://quotes.toscrape.com/page/{page}/"

    headers = {
        "User-Agent":"Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")

    # Extract quote details

    for quote in quotes:

        text = quote.find("span", class_="text").text

        author = quote.find("small", class_="author").text

        tags = quote.find_all("a", class_="tag")

        tag_names = ", ".join([tag.text for tag in tags])

        all_quotes.append(text)
        all_authors.append(author)
        all_tags.append(tag_names)

# Create DataFrame

df = pd.DataFrame({

    "Quote":all_quotes,

    "Author":all_authors,

    "Tags":all_tags

})

print("Total Records:",len(df))
```
